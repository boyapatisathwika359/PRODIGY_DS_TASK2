```python
import pandas as pd

df = pd.read_csv("titanic/train.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
df.drop("Cabin", axis=1, inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(x="Survived", data=df)
plt.title("Survival Count")
plt.show()
```

```python
sns.countplot(x="Sex", hue="Survived", data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
sns.countplot(x="Pclass", hue="Survived", data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Age"], bins=30, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Fare"], bins=30, kde=True)
plt.title("Fare Distribution")
plt.show()
```

```python
numeric_df = df.select_dtypes(include="number")

plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df[["Age", "Fare", "Pclass", "Survived"]], hue="Survived")
plt.show()
```

```python
print("Conclusion")
print("- Female passengers had a much higher survival rate than males.")
print("- First-class passengers had the highest survival rate.")
print("- Most passengers were between 20 and 40 years old.")
print("- Higher ticket fares were associated with better survival chances.")
print("- Passenger class and gender strongly influenced survival.")
```
